```python
### FILE PATHS 
ROOT_PATH_TE_FIMO = "/dors/capra_lab/users/abraha1/projects/transposable_elements/data/fimo_repeatmasker/fimo_output/individualTE_fimo_output"
TE_FIMO_FILE = {    
    "AluYa5":"fimo_AluYa5.txt",
    "AluYb8":"fimo_AluYb8.txt",
    "AluSp":"fimo_AluSp.txt",
    "AluY":"fimo_AluY.txt",
    "AluSc":"fimo_AluSc.txt",
    "AluSg":"fimo_AluSg.txt",
    "AluSq":"fimo_AluSq.txt",
    "AluSx":"fimo_AluSx.txt",
    "AluJb":"fimo_AluJb.txt",
    "AluJo":"fimo_AluJo.txt"} 

ROOT_PATH_ENHANCER_INTERACTION = "/dors/capra_lab/users/abraha1/projects/transposable_elements/data/tf_dynamics/alu_TF_cluster"
ENHANCER_INTERACTION_FILE = { 
    "AluJb":"AluJb_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluJo":"AluJo_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluSc":"AluSc_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluSg":"AluSg_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluSp":"AluSp_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluSq":"AluSq_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluSx":"AluSx_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluYa5":"AluYa5_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluYb8":"AluYb8_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv",
    "AluY":"AluY_FANTOM_enh_repeatmasker_TE_TF_intersection.tsv"}
```

```python
### LOAD FILE PATHS 
df = pd.read_csv(os.path.join(ROOT_PATH_TE_FIMO, TE_FIMO_FILE['AluJo']), sep="\t", header=None)
enh_df = pd.read_csv(os.path.join(ROOT_PATH_ENHANCER_INTERACTION, ENHANCER_INTERACTION_FILE['AluJo']), sep="\t")
enh_df['TE_sequence'] =  enh_df["chr_TE"].map(str) + ":" +  enh_df["start_TE"].map(str) + "-" +  enh_df["end_TE"].map(str)
```

```python
### CLEAN UP DATA
df.columns = ["TF", "TE", "TE_coordinates","TF_start_1based",
              "TF_end", "strand","motif_score","p_value",
              "motif_seq","q_value","num_bases_in_TE","num_bases_in_TFmotif"]

df = df.drop([ "motif_seq", "p_value", "strand","num_bases_in_TE","num_bases_in_TFmotif"], axis=1)
nodups_df = df.drop_duplicates(subset=['TF','TE_coordinates'], keep=False)
```

```python
### Split TF Col 
df['TF_name'], df['TF_quality'] = df['TF'].str.split('.H10MO.', 1).str
```

```python
### intersect enhancer-fantom overlap with TE overlap 
temp_enh_df = enh_df.loc[:,['TE_sequence','TFname']]
intersect_df = pd.merge(df, temp_enh_df, left_on=["TE_coordinates",'TF_name'], right_on=['TE_sequence','TFname'], how='left')
```

```python
### create boolean column for enhancer overlap for a given T 
intersect_df['enhancer_overlap'] = np.where(pd.isnull(intersect_df['TE_sequence']), False, True);
enhancer_overlap_key = intersect_df[ ['TE_coordinates', 'enhancer_overlap']]; 
enhancer_overlap_key.drop_duplicates(inplace=True);
```

```python
enhancer_overlap_key.shape
```

```python
TE_with_enhancer_overlap = enhancer_overlap_key[enhancer_overlap_key.enhancer_overlap == True]
```

```python
df.TE_coordinates.nunique()
```

```python
### Check Overlap between enhancer-TE-TF dataset and TE-TF dataset 
uniq_seq_enchTE = set(enh_df.TE_sequence.unique())
uniq_seq_df = set(df.TE_coordinates.unique())


print("Number of unique TEs in enhancer-te-tf file:   {}".format(len(uniq_seq_enchTE)))
print("Number of unique TEs in te-tf file:   {}".format(len(uniq_seq_df)))
print("Number of unique TEs in both of the above two files:   {}".format(len(uniq_seq_df.intersection(uniq_seq_enchTE))))
```

```python
### PRELIMINARY ANALYSIS 
## histogram of number of TF counts on each instance of TE 
raw_tf_counts = df.groupby(by='TE_coordinates').count()['TF']
unique_tf_counts = nodups_df.groupby(by='TE_coordinates').count()['TF']
TE_carrying_tf = nodups_df.groupby(by='TF').count()['TE_coordinates']
```

```python
## plot
fig, ax = plt.subplots(1,3)
_ = ax[0].hist(raw_tf_counts.values, alpha = 0.5, color = 'r')
_ = ax[0].set_xlabel('# of Raw TF Motifs')
_ = ax[0].set_ylabel('count')
_ = ax[0].set_title('# of TF per TE')

_ = ax[1].hist(unique_tf_counts.values, alpha = 0.5, color = 'g')
_ = ax[1].set_xlabel('# of Unique TF Motifs');
_ = ax[1].set_ylabel('count');
_ = ax[1].set_title('# of unique TF per TE')

_ = ax[2].hist(TE_carrying_tf.values, alpha = 0.5, color = 'b')
_ = ax[2].set_xlabel('# of TEs with A Specific TF');
_ = ax[2].set_ylabel('count');
_ = ax[2].set_title('# of TEs with a Specific TF')



plt.tight_layout()
plt.show()
```

NOTE: substantial number of rows are dropped when duplicate TE-TF pairs are removed.

```python
## GROUPBY

agg_func = {'motif_score':['count','mean','median']}
# nodups_df.groupby(by=['TF','TE_coordinates']).agg(agg_func).swaplevel('TE_coordinates','TF')
# gb_df = nodups_df.groupby(by=['TE_coordinates','TF']).agg(agg_func).reset_index()

gb_df = nodups_df.groupby(by=['TE_coordinates','TF']).agg(agg_func).unstack(fill_value=0)
count_gb_df = gb_df['motif_score']['count']
```

### Spectral Biclustering ### 


```python
from sklearn.cluster.bicluster import SpectralBiclustering
```

```python
data = gb_df.iloc[0:822,0:822].as_matrix()
```

```python
data.shape
```

```python
n_clusters = 10
model = SpectralBiclustering(n_clusters=n_clusters, method='bistochastic',random_state=0)
model.fit(data)

fit_data = data[np.argsort(model.row_labels_)]
fit_data = fit_data[:, np.argsort(model.column_labels_)]


plt.matshow(fit_data, cmap=plt.cm.Blues)
plt.title("After biclustering; rearranged to show biclusters")
plt.show()
```
